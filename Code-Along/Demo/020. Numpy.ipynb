# Numpy - Library to primarily handle arrays
- Statically typed (compiled in C, 50 times faster, element types can't be mixed or altered)
- Non resizable (can't append or remove elements)
- Much faster than python's dynamically typed lists
- Supports use of vector operations

## Create arrays

```python
import numpy as np

# In a list, each item is an object, with metadata and keeps track of the item type. This takes a lot of memory.
my_list = [5, True, "Hello"]
print(my_list[0])

# In this case the computer has to check what data type is each data type and what operations 
# are available before performing operations
print(my_list[0] + 3)

# In an array all items are of the same type. Type can be specified, but usually numpy interprets it well.
my_array = np.array([5,2,6,8], dtype=float)
print(my_array)

my_array.dtype
```

```python
a2 = np.random.random(20)
a3 = np.random.randint(10,20,10)
print(a2)
print(a3)
```

```python
a4 = np.zeros(10)
print(a4)
a5 = np.ones(10)
print(a5)
```

```python
a6 = np.linspace(0,10,100) # The final argument is the number of values to generate
print(a6)
a7 = np.arange(0,10,0.2) # Here the final argument is the step increment
a7[0] = 3 # Numpy arrays are changeable, you can change the value of individual items, but cannot add or remove items
a7
```

## Vectorized operations

```python
x = [2,7,4,6,3,6,67,8]

# Multiply by two. List comprehension or for loop.

x = [val*2 for val in x]
x
x*2
```

```python
# Numpy solution

x = np.array([2,7,4,6,3,6,67,8])
x

print(x/5)
print(np.sqrt(x)) # math.sqrt() can't operate on arrays, but only 1 value at a time. Most math operations exist in numpy.

x*2+1
```

```python
np.sin(x*2*np.pi) * 10 +1
```

```python
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
plt.plot(x,np.sin(x))
```

```python
a = np.array([3,6,67,8])
b = np.array([2,7,4,6])

print(a * b)

np.sqrt(a**2 + b**2)
```

## Array indexing and slicing
- Indexing and slicing works just as we are used to from Python lists.
- However, we can also use a list of indexs when indexing a numpy array.
- Boolean masks / boolean indexing can be used to filter through data in arrays.

```python
x = np.array([4, 14, 8, 12, 6, 12, 134, 16, 20]) 
x[-1]
```

```python
x[3:-2]
```

```python
# Getting multiple indexes is not possible with lists, but with numpy arrays.  Filtering using a list of indexes.
x[[5,1,0,4,3]]

#Boolean comparison to get the values you want. Boolean masking, that also can be created by Numpy.
x[[True,False,True,False,True,False,True,False,True]]
```

```python
# Boolean indexing

x[[False, False, False, True, True, False, False, True, False]]
```

```python
# Very useful for filtering
x[x>5]
```

## Aggregation methods
An aggregation method is a method that takes a list/array of values and by doing some calcultations returns a single value.

```python
x = np.array([4, 14, 8, 12, 6, 12, 134, 16, 20]) 
x
```

```python
np.sum(x)
```

```python
print(f"{x.sum() = }")
print(f"{x.min() = }")
print(f"{x.max() = }")
print(f"{x.mean() = }")
```

## Multi-dimensional arrays

```python
array_1d = np.array([4, 14, 8, 12, 6, 12, 134, 16, 20]) 
array_1d
```

```python
array_2d = np.array([[4, 14, 8, 12],[6, 12, 134, 16]]) 
array_2d
```

```python
# Specify size (and dimensionality) of array by passing a list
random_2d = np.random.randint(0,10,[5,4])
random_2d
```

```python
# Specify size (and dimensionality) of array by passing a list
random_3d = np.random.randint(0,10,[5,4,6])
random_3d
```

## Slicing of 2D array

```python
# Indexing in a 2D-array [x,y] position
print(random_2d)
random_2d[-1, 1]
```

```python
# Slicing in a 2D-array
# All rows, column 1
print(random_2d[:,1])

# First 3 rows, from second column 
print(random_2d[:3, 1:])

# You can also assign values with slices
random_2d[:3, 1:] = 0
```
