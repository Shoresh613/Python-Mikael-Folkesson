# Matplotlib

```python
import matplotlib.pyplot as plt

x = list(range(-10,11))
y = x
y2 = [value **2 for value in x]

print(x)
print(y)

plt.plot(x,y, label="y=x")
plt.plot(x,y2, label="y=x^2")
plt.legend()
```

```python
x = [0,-100,25,65,-323]
y = [0,3,7,3,9]

plt.title("My Graph")
plt.xlabel("X-values")
plt.ylabel("Y-values")

plt.axis([-50,80,2,8]) # To zoom in, default is otherwise often good

plt.plot(x,y, color="red", linestyle="dashed", marker="o")
```

## Scatter diagram

```python
x = list(range(-10,11))
y = [value **2 for value in x]

plt.scatter(x,y,marker="x")
plt.grid()
```

## Bars

```python
labels = ["Male", "Female"]
values = [15,25]

plt.bar(labels,values)
```

## Pie charts

```python
plt.pie(values, labels=labels)
```

# Sub plots

```python
import matplotlib.pyplot as plt

x = list(range(-10,11))
y = x
y2 = [value **2 for value in x]

fig, axs = plt.subplots(1,2)
axs[0].plot(x,y)
axs[1].plot(x,y2)

fig.set_facecolor("red")
axs[1].set_facecolor("blue")
axs[0].set_facecolor("green")
```
