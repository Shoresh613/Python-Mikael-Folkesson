````python
# b)

content = []
line=1
file_path = "..\\data\\test_result.txt"
print("Names sorted alphabetically:")

with open(file_path) as file:
    while line:
        line = file.readline().strip()
        content.append(line.strip())

content.sort()

for c in content:
    print(c)

with open(file_path,"a") as file: 
    file.writelines("\n\nAlphabetically Sorted: ")
with open(file_path,"a") as file: 
    file.writelines([" \n" + name for name in content if name!=""])
````

````python
# c)

grades = {
    "F": [0,19],
    "E": [20,29],
    "D": [30,39],
    "C": [40,49],
    "B": [50,59],
    "A": [60,69],
}

splitted_content = []
for c in content:
    if c!="":
        splitted_content.append(c.split(" "))
sorted_data = sorted(splitted_content, key=lambda x: x[-1])

with open(file_path,"a") as file: 
    file.writelines("\n\nGraded Results: ")

previous_grade = None

for result in sorted_data:
    score = int(result[-1])
    assigned_grade = None

    for grade, grade_range in grades.items():
        lower_bound, upper_bound = grade_range
        if lower_bound <= score <= upper_bound:
            assigned_grade = grade
            break

    if assigned_grade != previous_grade:
        with open(file_path,"a") as file: 
            file.writelines("\n " + "Grade: " + assigned_grade)
    with open(file_path,"a") as file: 
        result_str = " ".join(result)
        file.writelines("\n " + result_str)
    previous_grade = assigned_grade
````

## 3. National test (*)
Read in the file NPvt19Ma2A.txt and NPvt19Ma2C.txt (located in the data folder of this repo) in Python. Use **matplotlib** to plot pie charts for each grade categories in each file. (*)

[files]: https://github.com/kokchun/Programmering-med-Python-21/tree/main/Files

<details>

<summary>Hint </summary>

Use **subplot** in **matplotlib**

</details>
<br>
<details>

<summary>Answer </summary>

<img align="left" src="../Assets/NP.png" width="300"/>

</details>

````python
import matplotlib.pyplot as plt

line=1
categories_2A = []
with open("..\\data\\NPvt19Ma2A.txt","r") as file: 
    while line:
        line = file.readline().strip()
        if len(line) > 2:
            categories_2A.append(line.split(" "))
print(categories_2A)

line=1
categories_2C = []
with open("..\\data\\NPvt19Ma2C.txt","r") as file: 
    while line:
        line = file.readline().strip()
        if len(line) > 2:
            categories_2C.append(line.split(" "))
print(categories_2C)

grades2A = [float(x[1].removesuffix("%"))*100 for x in categories_2A]
share2A = [x[0] for x in categories_2A]

grades2C = [float(x[1].removesuffix("%"))*100 for x in categories_2C]
share2C = [x[0] for x in categories_2C]

fig, axs = plt.subplots(1,2)

axs[0].set_title('Grades 2A')
axs[1].set_title('Grades 2C')

axs[0].pie(grades2A, labels=share2A)
axs[1].pie(grades2C, labels=share2C)
plt.show()
````

## 5. Morse code (**)
Read in the file morse.txt, save it in a dictionary and create a function that lets the user input a message to get it translated to morse code. For example (**)

``` python
print(morse("SOS"))
print(morse("POKEMON"))
```
``` text
...---...
.------.-.------.
```


````python
morsedex = {' ': ' '}
line = 0
with open("..\\data\\morse.txt","r") as file: 
    for line in file:
        line = line.strip()
        if len(line) > 2:
            letter, code = line.split(": ", 1)
            morsedex.update({letter: code})

text = input("Enter text to translate to morse code: ")

for letter in text:
    print(morsedex[letter.upper()], end = '')
````
